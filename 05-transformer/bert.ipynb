```python
import torch
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")
```

```python
tokenizer = model[0].tokenizer
transformer = model[0].auto_model
```

```python
transformer.eval()
```

```python
sentence1 = "Dnes je vonku veľmi pekné počasie."
sentence2 = "Jablko má aj šušeň, aj stopku." # "Svieti slniečko, je ľahký vánok a príjemná teplota." #"Počasie je dnes naozaj krásne."
```

```python
token_ids1 = tokenizer(sentence1, padding=True, truncation=True, return_tensors="pt")
print(tokenizer.convert_ids_to_tokens(token_ids1["input_ids"][0]))
print(token_ids1["input_ids"].tolist())
print(tokenizer.decode(token_ids1["input_ids"][0]))
```

```python
token_ids2 = tokenizer(sentence2, padding=True, truncation=True, return_tensors="pt")
print(tokenizer.convert_ids_to_tokens(token_ids2["input_ids"][0]))
print(token_ids2["input_ids"].tolist())
print(tokenizer.decode(token_ids2["input_ids"][0]))
```

```python
# inference
with torch.no_grad():
    outputs1 = transformer(**token_ids1)
    outputs2 = transformer(**token_ids2)
```

```python
print(outputs1.last_hidden_state.shape)
print(outputs2.last_hidden_state.shape)
```

```python
# get CLS
embedding1 = outputs1.last_hidden_state[:, 0, :]
embedding2 = outputs2.last_hidden_state[:, 0, :]
```

```python
print(embedding1.shape)
print(embedding2.shape)
```

```python
similarity = util.cos_sim(embedding1, embedding2)
print("Cosine similarity:", similarity.item())
```

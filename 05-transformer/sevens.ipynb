```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('sevens.png')
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
binary = cv2.threshold(gray, 180, 255, cv2.THRESH_BINARY)[1]
binary = cv2.dilate(binary, None, iterations=7)
```

```python
# Connected component analysis
components = []
num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(binary, connectivity=8)
for i in range(1, num_labels):  # skip background (label 0)
    x = stats[i, cv2.CC_STAT_LEFT]
    y = stats[i, cv2.CC_STAT_TOP]
    w = stats[i, cv2.CC_STAT_WIDTH]
    h = stats[i, cv2.CC_STAT_HEIGHT]
    component = image[y:y+h, x:x+w]
    component[labels[y:y+h, x:x+w] != i] = (0,0,0)
    components.append(component)
```

```python
# display first 10 sevens
plt.figure(figsize=(16, 2))
for i in range(16):
    plt.subplot(1, 16, i+1)
    plt.imshow(components[i], cmap='gray')
plt.show()
```
